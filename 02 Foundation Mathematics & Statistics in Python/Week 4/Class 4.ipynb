```python
from scipy.stats import norm

# Load the Olympic athlete height data
olympic_data = pd.read_csv('olympic_athlete_database.csv')
athlete_heights = olympic_data['Height']

# Calculate mean and standard deviation of athlete heights
mean_height = athlete_heights.mean()
std_height = athlete_heights.std()

# Get the tallest and shortest student heights from the minor survey
tallest_student_height = Minorsurvey['length'].max()
shortest_student_height = Minorsurvey['length'].min()

# Calculate the probabilities
prob_taller_than_tallest = 1 - norm.cdf(tallest_student_height, mean_height, std_height)
prob_shorter_than_shortest = norm.cdf(shortest_student_height, mean_height, std_height)

# Calculate the percentage of Olympic athletes within the height range of the minor group
within_range = norm.cdf(tallest_student_height, mean_height, std_height) - norm.cdf(shortest_student_height, mean_height, std_height)
percentage_within_range = within_range * 100

print(f"Probability that an Olympic athlete is taller than the tallest student: {prob_taller_than_tallest:.4f}")
print(f"Probability that an Olympic athlete is shorter than the shortest student: {prob_shorter_than_shortest:.4f}")
print(f"Percentage of Olympic athletes within the height range of the minor group: {percentage_within_range:.2f}%")
```
