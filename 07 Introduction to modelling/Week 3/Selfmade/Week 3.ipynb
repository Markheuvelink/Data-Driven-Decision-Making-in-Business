```python
import pandas as pd

dataset_Facebook = pd.read_csv('dataset_Facebook - Copy.csv') 
dataset_Facebook_dummies = pd.get_dummies(dataset_Facebook['Payment'])
dataset_Facebook = pd.concat([dataset_Facebook,dataset_Facebook_dummies],axis=1)

dataset_Facebook.head()

import statsmodels.formula.api as sm
model1 = sm.ols ('like~Not_Paid', data=dataset_Facebook).fit()
print(model1.summary()) 
```

```python
import pandas as pd

dataset_insurance = pd.read_csv('insurance.csv') 

#1 variables
dataset_insurance_dummies_sex = pd.get_dummies(dataset_insurance['sex']) 
dataset_insurance_sex = pd.concat([dataset_insurance,dataset_insurance_dummies_sex],axis=1)

dataset_insurance_sex.head()
```

```python
import statsmodels.formula.api as sm

model2 = sm.ols (' charges ~ male', data=dataset_insurance_sex).fit()
print(model2.summary()) 
```

```python
#2 variables
dataset_insurance_dummies_both = pd.get_dummies(dataset_insurance[['sex','smoker']]) 
dataset_insurance_both = pd.concat([dataset_insurance,dataset_insurance_dummies_both],axis=1)

dataset_insurance_both.head()
```

```python
import statsmodels.formula.api as sm

model3 = sm.ols (' charges ~ sex_male+smoker_no', data=dataset_insurance_both).fit()
print(model3.summary()) 
```

```python
import pandas as pd

dataset_insurance = pd.read_csv('insurance.csv') 

dataset_insurance_dummies = pd.get_dummies(dataset_insurance[['sex','smoker']]) 
dataset_insurance = pd.concat([dataset_insurance,dataset_insurance_dummies],axis=1)

dataset_insurance.head()
```

```python
import statsmodels.formula.api as sm

model4 = sm.ols (' charges ~ sex_male', data=dataset_insurance).fit()

print(model4.summary()) 
```

```python
model5 = sm.ols (' charges ~ sex_male + smoker_yes', data=dataset_insurance).fit()
print(model5.summary())
```
