```python
from statsmodels.stats.outliers_influence import variance_inflation_factor as vif #import the VIF function
from statsmodels.tools.tools import add_constant #import the add_constant function
```

```python
#Exercise 1

import pandas as pd
WorldHappiness = pd.read_csv('World Happiness 2019.csv')
WorldHappiness.head()


```

```python
#Exercise 1a
import statsmodels.formula.api as sm

model1 = sm.ols('Happiness_Score ~ GDP_per_capita', data = WorldHappiness).fit()
print(model1.summary())
model2 = sm.ols('Happiness_Score ~ Social_support', data = WorldHappiness).fit()
print(model2.summary())
model3 = sm.ols('Happiness_Score ~ Healthy_life_expectancy', data = WorldHappiness).fit()
print(model3.summary())
model4 = sm.ols('Happiness_Score ~ Freedom_to_make_life_choices', data = WorldHappiness).fit()
print(model4.summary())
model5 = sm.ols('Happiness_Score ~ Generosity', data = WorldHappiness).fit()
print(model5.summary())
model6 = sm.ols('Happiness_Score ~ Perceptions_of_corruption', data = WorldHappiness).fit()
print(model6.summary())
```

```python
#Exercise 1b
X = WorldHappiness[['GDP_per_capita','Social_support','Healthy_life_expectancy','Freedom_to_make_life_choices','Generosity','Perceptions_of_corruption']].dropna() #drop missings
X = add_constant(X) #add a constant to the variables. VIF WON'T WORK CORRECTLY IF YOU DON'T DO THIS
#Create a dataframe that contains the VIF values
VIF_data = pd.DataFrame() #create a pandas dataframe
VIF_data["variable"] = X.columns #create a variable that simply lists the names of the variables you picked
VIF_data["VIF"] = [vif(X,i) 
                   for i in range(len(X.columns))] #a loop that creates vif values for each variable and then puts it in the VIF column
print(VIF_data) #print the results
```

```python
#Exercise 1c
model7 = sm.ols('Happiness_Score ~ GDP_per_capita+Social_support+Healthy_life_expectancy+Freedom_to_make_life_choices+Generosity+Perceptions_of_corruption', 
                data=WorldHappiness).fit()
print(model7.summary())
```

```python
#Exercise 3
import pandas as pd
GBvideos = pd.read_csv("GBvideos.csv")
GBvideos.head()

```

```python
#3b

#remove variables with no likes or dislikes
GBvideos = GBvideos[GBvideos.likes > 0] #interestingly, this one is not strictly necessary because there are no videos with > 0 dislikes and 0 likes
GBvideos = GBvideos[GBvideos.dislikes > 0]
        
#Remove rows which have blocked ratings, blocked comments, or have been removed
GBvideos = GBvideos[GBvideos.comments_disabled == False] #note that in the csv file it's spelled FALSE but in Python as False. This is because it is not a string variable but a boolean
GBvideos = GBvideos[GBvideos.ratings_disabled == False]
GBvideos = GBvideos[GBvideos.video_error_or_removed == False]
```

```python
#3c
#Create new variable
GBvideos = GBvideos.assign(like_percentage = GBvideos.likes/(GBvideos.likes+GBvideos.dislikes))

#check if there are no mistakes
GBvideos.head()
```

```python
#3d
from statsmodels.stats.outliers_influence import variance_inflation_factor as vif 
from statsmodels.tools.tools import add_constant 

X = GBvideos[['views','comment_count','like_percentage']].dropna() 
X = add_constant(X) #add a constant to the variables. VIF WON'T WORK CORRECTLY IF YOU DON'T DO THIS

#Create a dataframe that contains the VIF values
VIF_data = pd.DataFrame() #create a pandas dataframe
VIF_data["variable"] = X.columns #create a variable that simply lists the names of the variables you picked
VIF_data["VIF"] = [vif(X,i) 
                   for i in range(len(X.columns))] #a loop that creates vif values for each variable and then puts it in the VIF column
print(VIF_data) #print the results

```

```python
import statsmodels.formula.api as sm
model10 = sm.ols('like_percentage~comment_count+views', data=GBvideos).fit()
print(model10.summary())
```
