### Import data

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from matplotlib.ticker import FuncFormatter
import numpy as np
from matplotlib.ticker import MultipleLocator, FuncFormatter
import seaborn as sns



# Data for the bar graph
Data = pd.read_excel('NL_bicycle_sales.xlsx')
```

### Test

```python
# Plot the data
plt.figure(figsize=(10, 6))

# Plot each bicycle type with dots
plt.plot(Data['Year'], Data[' Electric bicycles'], label=' Electric bicycles', color='red', linewidth=2.5, marker='o')
plt.plot(Data['Year'], Data['Regular bicycles'], label='Regular bicycles', color='blue', linewidth=2.5, marker='o')
plt.plot(Data['Year'], Data[' Hybrid bicycles'], label=' Hybrid bicycles', color='gray', linewidth=1.5, linestyle='--')
plt.plot(Data['Year'], Data[' Kids bicycles'], label=' Kids bicycles', color='gray', linewidth=1.5, linestyle='--')
plt.plot(Data['Year'], Data[' Other bicycles'], label=' Other bicycles', color='gray', linewidth=1.5, linestyle='--')


# Add data labels at the end of each line
plt.text(Data['Year'].iloc[-1] + 0.2, Data[' Electric bicycles'].iloc[-1], 'Electric bicycles', color='red', fontsize=10)
plt.text(Data['Year'].iloc[-1] + 0.2, Data['Regular bicycles'].iloc[-1], 'Regular bicycles', color='blue', fontsize=10)
plt.text(Data['Year'].iloc[-1] + 0.2, Data[' Hybrid bicycles'].iloc[-1], 'Hybrid bicycles', color='gray', fontsize=10)
plt.text(Data['Year'].iloc[-1] + 0.2, Data[' Kids bicycles'].iloc[-1], 'Kids bicycles', color='gray', fontsize=10)
plt.text(Data['Year'].iloc[-1] + 0.2, Data[' Other bicycles'].iloc[-1], 'Other bicycles', color='gray', fontsize=10)

# Add labels, title, and legend
plt.xlabel('Year')
plt.ylabel('Proportion (%)')
plt.title('Bicycle Sales Over Time')

# Get current axis and display only the outside lines on the x and y axes
# Format the y-axis to show percentages
ax = plt.gca()
ax.yaxis.set_major_formatter(FuncFormatter(lambda y, _: f'{y*100:.0f}%'))
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Set y-axis limits to start at 0% and end at 60%
plt.ylim(0, 0.6)

# Show each year on the x-axis
plt.xticks(Data['Year'], rotation=45)  # Rotate the labels for better readability

# Show the plot
plt.grid(True)
plt.show()
```

### Result

```python
# Plot the data
plt.figure(figsize=(6, 6))

# Helper function to add percentage labels above dots
def add_percentage_label(x, y, color):
    for i in x.index:
        plt.text(x[i], y[i] + 0.02, f'{y[i]*100:.0f}%', color=color, fontsize=9, ha='center')

# Plot each bicycle type with only the first and last two dots
for column, color, label in [
    (' Electric bicycles', 'orange', 'Electric bicycles'),
    ('Regular bicycles', 'blue', 'Regular bicycles')
]:
    x = Data['Year']
    y = Data[column]
    plt.plot(x, y, label=label, color=color, linewidth=2.5, marker='o', markevery=[0, -2, -1])
    add_percentage_label(x.iloc[[0, -2, -1]], y.iloc[[0, -2, -1]], color)

# Plot other bicycle types without dots
for column, label in [
    (' Hybrid bicycles', 'Hybrid bicycles'),
    (' Kids bicycles', 'Kids bicycles'),
    (' Other bicycles', 'Other bicycles')
]:
    plt.plot(Data['Year'], Data[column], label=label, color='gray', linewidth=1.5, linestyle='--')

# Add data labels at the end of each line
plt.text(Data['Year'].iloc[-1] + 0.2, Data[' Electric bicycles'].iloc[-1], 'Electric bicycles', color='orange', fontsize=10)
plt.text(Data['Year'].iloc[-1] + 0.2, Data['Regular bicycles'].iloc[-1], 'Regular bicycles', color='blue', fontsize=10)
plt.text(Data['Year'].iloc[-1] + 0.2, Data[' Hybrid bicycles'].iloc[-1], 'Hybrid bicycles', color='gray', fontsize=10)
plt.text(Data['Year'].iloc[-1] + 0.2, Data[' Kids bicycles'].iloc[-1], 'Kids bicycles', color='gray', fontsize=10)
plt.text(Data['Year'].iloc[-1] + 0.2, Data[' Other bicycles'].iloc[-1], 'Other bicycles', color='gray', fontsize=10)


# Add labels, title, and legend
plt.xlabel('Year')
plt.ylabel('Proportion (%)')
plt.title('The growth of electric bicycles')


# Get current axis and display only the outside lines on the x and y axes
# Format the y-axis to show percentages
ax = plt.gca()
ax.yaxis.set_major_formatter(FuncFormatter(lambda y, _: f'{y*100:.0f}%'))
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Set y-axis limits to start at 0% and end at 60%
plt.ylim(0, 1)

# Show each year on the x-axis
plt.xticks(Data['Year'], rotation=45)  # Rotate the labels for better readability

# Show the plot
plt.grid(False)
plt.show()
```
