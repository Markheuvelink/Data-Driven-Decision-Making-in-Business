## 1. Raw Material Stock Levels Over Time

```python
import pandas as pd
import matplotlib.pyplot as plt

# Example: Raw material stock levels over time
raw_material_data = pd.read_excel('Bean there done that data.xlsx', sheet_name='Inventory (raw material)')

# Group by month and product type
stock_levels = raw_material_data.groupby(['Month', 'Product'])['Stock level'].sum().unstack()

# Plot stock levels
stock_levels.plot(kind='line', figsize=(10, 6), marker='o')
plt.title('Raw Material Stock Levels Over Time')
plt.xlabel('Month')
plt.ylabel('Stock Level')
plt.legend(title='Product', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
# Display the plot
plt.show()

# Add a note for better understanding
print("The line chart above shows the stock levels of raw materials over time, grouped by product.")
```

## 2. Finished Product Stock vs. Demand

```python
# Example: Finished product stock vs. e-commerce demand
finished_product_data = pd.read_excel('Bean there done that data.xlsx', sheet_name='Inventory (finished product)')
ecom_sales_data = pd.read_excel('Bean there done that data.xlsx', sheet_name='E-com sales orders by product')

# Ensure the 'Month' column in finished_product_data is in datetime format
finished_product_data['Month'] = pd.to_datetime(finished_product_data['Month'], format='%B')

# Aggregate finished product stock by month
finished_stock = finished_product_data.groupby('Month')['Stock level'].sum()

# Ensure the 'Order date' column in ecom_sales_data is in datetime format
ecom_sales_data['Order date'] = pd.to_datetime(ecom_sales_data['Order date'])

# Aggregate e-commerce demand by month
ecom_sales_data['Month'] = ecom_sales_data['Order date'].dt.to_period('M').dt.to_timestamp()
ecom_demand = ecom_sales_data.groupby('Month')['Quantity'].sum()

# Combine into one DataFrame
comparison = pd.DataFrame({'Stock Level': finished_stock, 'E-com Demand': ecom_demand})

# Plot
comparison.plot(kind='line', figsize=(10, 6), marker='o')
plt.title('Finished Product Stock vs. E-commerce Demand')
plt.xlabel('Month')
plt.ylabel('Quantity')
plt.legend(loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## 3. Roasting Quality vs. Finished Product Stock

```python
# Example: Roasting quality vs. finished product stock
roasting_data = pd.read_excel('Bean there done that data.xlsx', sheet_name='Roasting')

# Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(roasting_data['Quality Score'], roasting_data['Batch Size'], alpha=0.7)
plt.title('Roasting Quality vs. Batch Size')
plt.xlabel('Quality Score')
plt.ylabel('Batch Size')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## 4. E-com Sales Orders by Product ↔ B2B Orders

```python
# Example: Top-selling products in e-commerce
top_products = ecom_sales_data.groupby('Product')['Quantity Sold'].sum().sort_values(ascending=False)

# Bar chart
top_products.plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.title('Top-Selling Products (E-commerce)')
plt.xlabel('Product')
plt.ylabel('Quantity Sold')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 5. Inventory (Raw Material) ↔ Inventory (Finished Product)

```python
# Example: B2B order volume by client
b2b_orders_data = pd.read_excel('Bean there done that data.xlsx', sheet_name='B2B orders')

# Aggregate data
b2b_volume = b2b_orders_data.groupby('Client')['Order Volume'].sum()

# Bar chart
b2b_volume.plot(kind='bar', figsize=(10, 6), color='orange')
plt.title('B2B Order Volume by Client')
plt.xlabel('Client')
plt.ylabel('Order Volume')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 6. Roasting ↔ B2B Orders

```python
# Example: Stock allocation between B2B and E-commerce
b2b_stock = b2b_orders_data.groupby('Date')['Order Volume'].sum()
ecom_stock = ecom_sales_data.groupby('Date')['Quantity Sold'].sum()

# Combine into one DataFrame
stock_allocation = pd.DataFrame({'B2B Stock': b2b_stock, 'E-com Stock': ecom_stock})

# Stacked bar chart
stock_allocation.plot(kind='bar', stacked=True, figsize=(10, 6), color=['orange', 'skyblue'])
plt.title('Stock Allocation: B2B vs. E-commerce')
plt.xlabel('Date')
plt.ylabel('Stock Volume')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

## 7. E-com Sales Orders by Product ↔ Inventory (Raw Material)

```python
# Example: Supplier performance (raw material delivery times)
supplier_performance = raw_material_data.groupby('Supplier')['Delivery Time (Days)'].mean()

# Bar chart
supplier_performance.plot(kind='bar', figsize=(10, 6), color='green')
plt.title('Supplier Performance (Average Delivery Time)')
plt.xlabel('Supplier')
plt.ylabel('Delivery Time (Days)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 8. Roasting Downtime

```python
# Example: Roasting downtime analysis
downtime_data = roasting_data.groupby('Date')['Downtime (Hours)'].sum()

# Line chart
downtime_data.plot(kind='line', figsize=(10, 6), marker='o', color='red')
plt.title('Roasting Downtime Over Time')
plt.xlabel('Date')
plt.ylabel('Downtime (Hours)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
